```python
# Ruta de la carpeta que contiene los archivos shp
carpeta = 'TIF4Stats/Stats_Shp'

# Obtener la lista de archivos shp en la carpeta
archivos_shp = [archivo for archivo in os.listdir(carpeta) if archivo.endswith('.shp')]

# Lista para almacenar los geodataframes
geodataframes = []

# Iterar sobre cada archivo shp
for archivo in archivos_shp:
    # Ruta completa del archivo shp
    ruta_archivo = os.path.join(carpeta, archivo)
    
    # Leer el archivo shp y convertirlo en un geodataframe
    gdf = gpd.read_file(ruta_archivo)
    
    # Aplicar la función sort_values('Class_9')
    gdf_sorted = gdf.sort_values('Class_9', ascending= False)
    
    # Agregar la columna "Calif" con valor 1 en las filas donde "Class_9" es distinto de 0
    gdf_sorted.loc[gdf_sorted['Class_9'] != 0, 'Calif'] = 1
    
    # Agregar la columna "Aptitud" y asignar 'Aptitud Muy Alta' en las filas donde "Calif" es 1
    gdf_sorted.loc[gdf_sorted['Calif'] == 1, 'Aptitud'] = 'Aptitud Muy Alta'
    
    # Agregar la columna "Calif" con valor 2 en las filas donde "Class_8" es distinto de 0
    gdf_sorted.loc[gdf_sorted['Class_8'] != 0, 'Calif'] = 2
    # Asignar 'Aptitud Alta' en las filas donde "Calif" es 2
    gdf_sorted.loc[gdf_sorted['Calif'] == 2, 'Aptitud'] = 'Aptitud Alta'
    
    # Agregar la columna "Calif" con valor 3 en las filas donde "Class_7" es distinto de 0
    gdf_sorted.loc[gdf_sorted['Class_7'] != 0, 'Calif'] = 3
    # Asignar 'Aptitud Alta Baja' en las filas donde "Calif" es 3
    gdf_sorted.loc[gdf_sorted['Calif'] == 3, 'Aptitud'] = 'Aptitud Alta Baja'
    
    # Agregar la columna "Calif" con valor 4 en las filas donde "Class_6" es distinto de 0
    gdf_sorted.loc[gdf_sorted['Class_6'] != 0, 'Calif'] = 4
    # Asignar 'Aptitud Media Alta' en las filas donde "Calif" es 4
    gdf_sorted.loc[gdf_sorted['Calif'] == 4, 'Aptitud'] = 'Aptitud Media Alta'
    
    # Agregar la columna "Calif" con valor 5 en las filas donde "Class_5" es distinto de 0
    gdf_sorted.loc[gdf_sorted['Class_5'] != 0, 'Calif'] = 5
    # Asignar 'Aptitud Media' en las filas donde "Calif" es 5
    gdf_sorted.loc[gdf_sorted['Calif'] == 5, 'Aptitud'] = 'Aptitud Media'
    
    # Agregar la columna "Calif" con valor 6 en las filas donde "Class_4" es distinto de 0
    gdf_sorted.loc[gdf_sorted['Class_4'] != 0, 'Calif'] = 6
    # Asignar 'Aptitud Media Baja' en las filas donde "Calif" es 6
    gdf_sorted.loc[gdf_sorted['Calif'] == 6, 'Aptitud'] = 'Aptitud Media Baja'
    
    # Agregar la columna "Calif" con valor 7 en las filas donde "Class_3" es distinto de 0
    gdf_sorted.loc[gdf_sorted['Class_3'] != 0, 'Calif'] = 7
    # Asignar 'Aptitud Baja' en las filas donde "Calif" es 7
    gdf_sorted.loc[gdf_sorted['Calif'] == 7, 'Aptitud'] = 'Aptitud Baja'
    
    # Agregar la columna "Calif" con valor 8 en las filas donde "Class_2" es distinto de 0
    #gdf_sorted.loc[gdf_sorted['Class_2'] != 0, 'Calif'] = 8
    # Asignar 'Aptitud Muy Baja' en las filas donde "Calif" es 8
    #gdf_sorted.loc[gdf_sorted['Calif'] == 8, 'Aptitud'] = 'Aptitud Muy Baja'
    
    # Agregar la columna "Calif" con valor 9 en las filas donde "Class_1" es distinto de 0
    #gdf_sorted.loc[gdf_sorted['Class_1'] != 0, 'Calif'] = 9
    # Asignar 'No Apto' en las filas donde "Calif" es 9
    #gdf_sorted.loc[gdf_sorted['Calif'] == 9, 'Aptitud'] = 'No Apto'
    
    # Agregar el geodataframe modificado a la lista
    geodataframes.append(gdf_sorted)
```
